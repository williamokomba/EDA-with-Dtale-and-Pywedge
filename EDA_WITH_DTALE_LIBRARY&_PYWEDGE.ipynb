### **Visualization using PyWedge Dashboard**

```python
#Installing the package
#!pip install pywedge # installing pywedge package

#importing module
import pywedge as pw # installing pywedge
import pandas as pd
```

```python
#loading the dataset ans creating tha dataframe
data=pd.read_csv("/content/sample_data/california_housing_train.csv")
data.head()
```

#creating the dashboard

* C is the variable that you don't want be part of the analysis i.e id column(varaible that need to be dropped). If nothing has to be dropped then C=None.
* y is the target variable. 

```python
#creating the dashboard
mc= pw.Pywedge_Charts(data, c=None, y= "median_house_value")

charts= mc.make_charts()
```
